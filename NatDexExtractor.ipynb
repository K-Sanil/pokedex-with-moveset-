```python
import pandas as pd 
import requests 
from bs4 import BeautifulSoup 
import numpy as np
```

```python
webPage = requests.get("https://pokemondb.net/pokedex/national").text
```

```python
webPage
```

```python
soup = BeautifulSoup(webPage, 'lxml')
```

## Listing out all the generation available to us

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

## Checkin if the specific class is including all pokemons or not 

```python
len(soup.find_all('span' , class_ = 'infocard-lg-data text-muted'))
```

```python
pokemon = soup.find_all('span' , class_ = 'infocard-lg-data text-muted')
len(pokemon)
```

## Extracting Name of the pokemon and type

```python
pokedexNo = []
name = []
type = []

for i in pokemon:
    pokedexNo.append(i.find_all('small')[0].text.strip())
    name.append(i.find('a' , class_ = 'ent-name').text.strip())
    type.append(i.find_all('small')[1].text.strip())

df = pd.DataFrame({'NatDex#': pokedexNo,
                    'name':name,
                   'type':type    
})
```
