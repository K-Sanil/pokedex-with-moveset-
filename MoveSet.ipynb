## Extracting GEN 9 movesets of Bulbasaur

```python
webPage = requests.get('https://pokemondb.net/pokedex/bulbasaur/moves/9').text
```

```python
soup = BeautifulSoup(webPage , 'lxml')
```

```python
len(soup.find_all('table' , class_ = 'data-table'))
```

```python
soup.find_all('table' , class_ = 'data-table')[0].text
```

```python
soup.find_all('tbody')[0].text
```

```python
moveTable = soup.find_all('tbody')[0]
```

```python
lvlUPmoves = moveTable.find_all('tr')
```

```python
lvl = []
moveName = []
type = []
category = []
power = []
accuracy = []

for i in lvlUPmoves:
    try:
        lvl.append(i.find_all('td')[0].text.strip())
    except:
        lvl.append(np.nan)
    try:
         moveName.append(i.find_all('td')[1].text.strip())
    except:
        moveName.append(np.nan)
    try:
        type.append(i.find_all('td')[2].text.strip())
    except:
        type.append(np.nan)

## extracting name from alt tag of image
    try:
        imgT = i.find('img')
        category.append(imgT['alt'])
    except:
        category.append(np.nan)

    
    try:
        power.append(i.find_all('td')[4].text.strip())
    except:
        power.append(np.nan)
    try:
        accuracy.append(i.find_all('td')[5].text.strip())
    except:
        accuracy.append(np.nan)

df = pd.DataFrame({'level': lvl,
                   'move': moveName,
                   'type': type,
                   'category': category,
                   'power': power,
                   'accuracy': accuracy,
                   })
```

## Creating Data base for all Pokemon in the list

(applied try and except on moveTable as some pokemons are not available in GEN 9 so no moveset , those pokemon will be skipped )

```python
final = pd.DataFrame()

for j in pokeName:
    webPage = requests.get('https://pokemondb.net/pokedex/{}/moves/9'.format(j)).text
    soup = BeautifulSoup(webPage , 'lxml')
    try:
        moveTable = soup.find_all('tbody')[0]
    except:
        continue
    lvlUPmoves = moveTable.find_all('tr')

    name = []
    lvl = []
    moveName = []
    type = []
    category = []
    power = []
    accuracy = []

    for i in lvlUPmoves:
        try:
            name.append(j)
        except:
            name.append(np.nan)
        try:
            lvl.append(i.find_all('td')[0].text.strip())
        except:
            lvl.append(np.nan)
        try:
             moveName.append(i.find_all('td')[1].text.strip())
        except:
            moveName.append(np.nan)
        try:
            type.append(i.find_all('td')[2].text.strip())
        except:
            type.append(np.nan)

## extracting name from alt tag of image
        try:
            imgT = i.find('img')
            category.append(imgT['alt'])
        except:
            category.append(np.nan)

    
        try:
            power.append(i.find_all('td')[4].text.strip())
        except:
            power.append(np.nan)
        try:
            accuracy.append(i.find_all('td')[5].text.strip())
        except:
            accuracy.append(np.nan)


    df = pd.DataFrame({'name': name,
                          'levelLearned': lvl,
                           'move': moveName,
                          'type': type,
                          'category': category,
                          'power': power,
                          'accuracy': accuracy})

    final = pd.concat([final , df] , ignore_index = True) 
    
```
